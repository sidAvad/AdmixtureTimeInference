```python
import sys
sys.path.insert(1, '/Users/siddharth/cbsu_rsync/AncestryInferenceV2')

from AncestryInferenceV2 import dataViz as dvanc

importlib.reload(dvanc)

#Compute sums and counts for all different tract lengths
tractlendf['hom0_sum'] = tractlendf['hom_0'].apply(lambda x: dvanc.sum_tract_lengths(x))
tractlendf['hom1_sum'] = tractlendf['hom_1'].apply(lambda x: dvanc.sum_tract_lengths(x))
tractlendf['het_sum'] = tractlendf['het'].apply(lambda x: dvanc.sum_tract_lengths(x))
tractlendf['total_sum'] = tractlendf['merged_tract_lengths'].apply(lambda x: dvanc.sum_tract_lengths(x))

tractlendf['hom0_count'] = tractlendf['hom_0'].apply(lambda x: len(x))
tractlendf['hom1_count'] = tractlendf['hom_1'].apply(lambda x: len(x))
tractlendf['het_count'] = tractlendf['het'].apply(lambda x: len(x))

tractlendf['hom0_prop'] = tractlendf.apply(lambda x: x['hom0_sum']/x['total_sum'], axis=1)
tractlendf['het_prop'] = tractlendf.apply(lambda x: x['het_sum']/x['total_sum'], axis=1)

tractlendf

#Get inferred p1 and p2 first as well as true p1 and p2 columns ( for ancestry label 0)

tractlendf['p1_inf'] = tractlendf.apply(lambda x: dvanc.do_inference(x['hom0_prop'], x['het_prop'], x.index, x['admixtype'])['p1'], axis=1)
tractlendf['p2_inf'] = tractlendf.apply(lambda x: dvanc.do_inference(x['hom0_prop'], x['het_prop'], x.index, x['admixtype'])['p2'], axis=1)


importlib.reload(dv)

#Infer t1 and t2 using get_simult_mle_df
tractlendf['t1'] = tractlendf.apply(lambda x: dv.get_simult_mle_df(x['hom_0'], x['hom_1'], x['p1_inf'][0], x['p2_inf'][0], sum(x['hom0_status']), sum(x['hom1_status']))[0], axis=1)
tractlendf['t2'] = tractlendf.apply(lambda x: dv.get_simult_mle_df(x['hom_0'], x['hom_1'], x['p1_inf'][0], x['p2_inf'][0], sum(x['hom0_status']), sum(x['hom1_status']))[1], axis=1)
```
